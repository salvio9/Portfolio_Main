```python
import random


class Cientist:
    def __init__(self,name,apt):
        self.name = name
        self.apt = apt
        self.project = None
        self.apt_project = None

    def __str__(self):
        return print(self.project)


apts =  [ [70 ,65, 55 ,50, 90 ,67 ,80 ,62 ,100 ,85],[74 ,83 ,54 ,60 ,100 ,75 ,50 ,76 ,82 ,45],[71 ,87 ,66 ,58 ,74 ,81 ,48 ,52 ,100, 64
            ],[50, 89, 78, 63, 51, 40, 48, 100, 68, 75],[100 ,66 ,83, 77, 54, 58 ,93, 89, 53, 48],[80,55 ,70 ,65 ,94 ,47 ,60 ,88 ,73 ,60],
          [87 ,63 ,90 ,79 ,47 ,77 ,90 ,76, 85, 90],[67, 95 ,100, 40, 70, 54 ,70 ,100, 65 ,57],[90 ,45 ,88 ,48 ,65 ,68 ,80 ,46 ,71 ,67],
          [67 ,77 ,50, 60, 100, 70, 80, 60, 65 ,70]]






```

```python


def get_initial_solution(cientists):
    cientists_use = []
    for i in cientists:
        cientists_use.append(i)

    cientists_initial = []


    for i in cientists:
        cientists_initial.append(i)



    available = []

    while len(available) != len(cientists_use):
        cien = random.choice(cientists_initial)
        cientists_initial.remove(cien)

        for i in cientists_use:

            if i.name == cien.name:
                apts = []
                for k in i.apt:
                    apts.append(k)

                apts.sort(reverse=True)
                for j in apts:

                    if i.apt.index(j) not in available:
                        i.project = i.apt.index(j)+1
                        i.apt_project = j
                        available.append(i.apt.index(j))
                        break

    return cientists_use
```

```python
def tabu_search(tabu_n,iter,cientists_trab):
    sums = []

    cientists = get_initial_solution(cientists_trab)

    sum_ini = 0
    for i in cientists:
        sum_ini = sum_ini + i.apt_project

    best_comb = cientists

    sum_best = sum_ini

    tabu = []
    for i in range (0,tabu_n):
        tabu.append(i)


    t = 0

    while t <= iter:
        profit = -130000
        trade = None
        not_tabu = True
        for i in range(0,len(cientists)):
            if not_tabu == False:
                break

            for j in range(0,len(cientists)):
                if i != j:



                    if cientists[i].apt[cientists[j].project-1] + cientists[j].apt[cientists[i].project-1] > profit:
                        if [i,j] in tabu or [j,i] in tabu:

                            su = 0
                            for s in cientists:

                                if s.name == cientists[i].name:
                                    su = su + cientists[i].apt[cientists[j].project-1]
                                elif s.name == cientists[j].name:
                                    su = su + cientists[j].apt[cientists[i].project-1]
                                else:
                                    su = su + s.apt_project

                            if su > sum_best:
                                not_tabu = False
                                trade = [i,j]
                                break


                        else:
                            profit = cientists[i].apt[cientists[j].project-1] + cientists[j].apt[cientists[i].project-1]
                            trade = [i,j]



        pro_i = cientists[trade[0]].project
        cientists[trade[0]].project = cientists[trade[1]].project
        cientists[trade[1]].project = pro_i
        cientists[trade[0]].apt_project = cientists[trade[0]].apt[cientists[trade[0]].project - 1]
        cientists[trade[1]].apt_project = cientists[trade[1]].apt[cientists[trade[1]].project - 1]

        total = 0
        comb = []
        spend = []

        for i in cientists:
            total = total + i.apt_project
            spend.append(i.apt_project)
            comb.append(i.project)






        if total > sum_best:
            total = 0
            comb = []


            for i in cientists:
                total = total + i.apt_project

                a = Cientist(i.name,i.apt)
                a.project = i.project
                a.apt_project = i.apt_project

                comb.append(a)


            if total > sum_best:

                sums.append(total)
                sum_best = total
                best_comb = comb






        del tabu[-1]
        tabu.insert(0,trade)


        t = t + 1


    return best_comb
```

```python
def tabu_results(tabu_n,iter):
    cientists_trab = []

    for i in range(0,len(apts)):
        cientists_trab.append(Cientist(i+1,apts[i]))

    results = tabu_search(tabu_n,iter,cientists_trab)

    final = {}
    sum = 0

    for i in results:
        final[i.name] = i.project
        sum = sum + i.apt_project

    print(final)
    print(sum)

    return results
```
